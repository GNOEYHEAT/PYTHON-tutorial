```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pylab as plt
```

# Pandas의 시각화 기능 

```python
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                  index=pd.date_range('1/1/2018', periods=100),
                  columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1.plot()
plt.title("Pandas Plot Example")
plt.xlabel("time")
plt.ylabel("Data")
plt.show()
```

```python
iris = sns.load_dataset("iris") # 붓꽃 데이터
titanic = sns.load_dataset("titanic") # 타이타닉호 데이터

iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title("iris length visualization")
plt.xlabel("Data")
plt.ylabel("iris length")
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.title("Iris Bar Plot")
plt.xlabel("Data")
plt.ylabel("Feature Value")
plt.ylim(0, 7)
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = "feature"
df2
```

```python
df2.plot.bar(rot=0)
plt.title("Feature mean")
plt.xlabel("mean")
plt.ylabel("type")
plt.ylim(0, 8)
plt.show()
```

```python
df2.T.plot.bar(rot=0)
plt.title("Feature mean")
plt.xlabel("Feature")
plt.ylabel("mean")
plt.show()
```

```python
df3 = titanic.pclass.value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title("ratio")
plt.axis('equal')
plt.show()
```

```python
iris.plot.hist()
plt.title("Feature Histogram")
plt.xlabel("data value")
plt.show()
```

```python
iris.plot.kde()
plt.title("Feature Kernel Density Plot")
plt.xlabel("data value")
plt.show()
```

```python
iris.plot.box()
plt.title("Feature Box Plot")
plt.xlabel("Feature")
plt.ylabel("data value")
plt.show()
```

```python
iris.boxplot(by='species')
plt.tight_layout(pad=3, h_pad=1)
plt.title("Feature Box Plot")
plt.show()
```
