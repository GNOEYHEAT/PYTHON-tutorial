```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

np.random.seed(1011)

N = [1, 2, 10] #  

for _ in N:
    
    X = np.random.rand(5000, _) # X ~ uniform distribution(0, 1)
    Xbar = (X.mean(axis=1) - 0.5) * np.sqrt(12 * _) #  

    sns.distplot(Xbar, kde=True)
    plt.title("N = " + str(_))
    plt.show()

    sp.stats.probplot(Xbar, plot=plt)
    plt.show()
```
