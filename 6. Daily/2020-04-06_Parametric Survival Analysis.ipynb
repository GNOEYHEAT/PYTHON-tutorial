# Parametric Survival Analysis

```python
import numpy as np
import pandas as pd

from lifelines import ExponentialFitter, GeneralizedGammaFitter, WeibullFitter, LogNormalFitter
from lifelines.plotting import qq_plot
from lifelines.datasets import load_waltons

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_waltons = load_waltons()

df_waltons.info()
df_waltons.head()
```

```python
T = df_waltons['T']
E = df_waltons['E']

# timeline = np.linspace(0, 100, 200)

exf = ExponentialFitter().fit(T, E, label='ExponentalFitter')
gg = GeneralizedGammaFitter().fit(T, E, label='GeneralizedGammaFitter')
wbf = WeibullFitter().fit(T, E, label='WeibullFitter')
lnf = LogNormalFitter().fit(T, E, label='LogNormalFitter')
```

#### Gamma parameter
   * When λ=1 and σ=1, then the data is Exponential.
   * When λ=1 then the data is Weibull.
   * When σ=λ then the data is Gamma.
   * When λ=0 then the data is Log-Normal.

## 1. Exponential Distribution

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 3))

exf.print_summary()

exf.plot_survival_function(ax=axes[0])
exf.plot_cumulative_hazard(ax=axes[1])
qq_plot(exf, ax=axes[2])

plt.show()
```

## 2. Gamma distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 3))

gg.print_summary()

gg.plot_survival_function(ax=axes[0])
gg.plot_cumulative_hazard(ax=axes[1])

plt.show()
```

## 3. Weibull distribution

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 3))

wbf.print_summary()

wbf.plot_survival_function(ax=axes[0])
wbf.plot_cumulative_hazard(ax=axes[1])
qq_plot(wbf, ax=axes[2])

plt.show()
```

## 4. Log-Normal distribution

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 3))

lnf.print_summary()

lnf.plot_survival_function(ax=axes[0])
lnf.plot_cumulative_hazard(ax=axes[1])
qq_plot(lnf, ax=axes[2])

plt.show()
```
